```
import apache_beam as beam
```

```
def return_kv_tuple(element):
  return (element[0],element[1:])

with beam.Pipeline() as p1:

  dept_records = (
      p1
      |'Read dept records' >> beam.io.ReadFromText('dept_data.txt')
      |'Split lines to records' >> beam.Map(lambda lines : lines.split(','))
      |'call return tuple' >> beam.Map(return_kv_tuple)
      #|'Write dept' >> beam.io.WriteToText('data/dept')
  )

  loc_records = (
      p1
      |'Read location records' >> beam.io.ReadFromText('location.txt')
      |'Split loc records' >> beam.Map(lambda lines : lines.split(','))
      |'Create a tuple' >> beam.Map(return_kv_tuple)
      #|'Write loc' >> beam.io.WriteToText('data/loc')
  )

  joined_records = ({'dept': dept_records, 'loc':loc_records}
      |beam.CoGroupByKey()
      |'Write dept' >> beam.io.WriteToText('data/combined')
  )

     
```
